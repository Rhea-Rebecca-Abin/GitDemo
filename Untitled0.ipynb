```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 3 * x**2

# Generate x values
x_values = np.linspace(-1, 3, 100)

# Generate y values
y_values = f(x_values)

# Plot the function
plt.plot(x_values, y_values, label='y = 3x^2')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = 3x^2')

# Add a grid
plt.grid(True)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
